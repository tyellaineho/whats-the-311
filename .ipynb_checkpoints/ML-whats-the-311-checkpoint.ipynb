```python
import pandas as pd
from sodapy import Socrata
```

```python
# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
# client = Socrata("https://data.cityofnewyork.us", None)

# Example authenticated client (needed for non-public datasets):
client = Socrata("data.cityofnewyork.us",
                 YourAccessToken,
                 username= YourEmail,
                 password= YourPassword)

client.timeout = 300

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("wewp-mm3p", where="date_time between '2019-01-01T12:00:00' and '2019-12-31T23:59:00'", limit = 5568370)

# $where=date_time%20between%20%272019-01-10T00:00:00%27%20and%20%272019-12-31T23:59:00%27

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

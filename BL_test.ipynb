```python
data = pd.read_csv("311-complaints.csv")
```

```python
data
```

```python
data['complaint'] = data['inquiry_name'].str.contains('Complaint')
```

```python
data.complaint = data.complaint.astype(int)
```

```python
data
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = data.sample(n = 100000) 
```

```python
data['year'], data['t_column'] = data['date'].str.split('T', 1).str
```

```python
data['year'], data['month'] = data['year'].str.split('-', 1).str
```

```python
data['month'], data['calendar_day'] = data['month'].str.split('-', 1).str
data['date'], data['t_column_2'] = data['date'].str.split('T', 1).str
```

```python
data
```

```python
organizedcomplaint_df = data[["agency",
                              "agency_name",
                              "brief_description",
                              "call_resolution",
                              "date",
                              "inquiry_name",
                              "time",
                              "complaint",
                              "year",
                              "month",
                              "calendar_day"]]
```

```python
organizedcomplaint_df
```

```python
organizedcomplaint_df['time_new'] = pd.to_datetime(organizedcomplaint_df['time']).dt.strftime('%H:%M:%S')
```

```python
organizedcomplaint_df
```

```python
organizedcomplaint_df['time_hour'] = organizedcomplaint_df['time_new'].str.split(':').str[0]
```

```python
organizedcomplaint_df
```

```python
pd.value_counts(organizedcomplaint_df['year']).plot.bar()
```

```python
from sklearn.model_selection import train_test_split
```

```python
organizedcomplaint_df
```

```python
data_for_x = organizedcomplaint_df[['time_hour','month']]
data_for_y = organizedcomplaint_df[['complaint']]
```

```python
X = data_for_x.values
y = data_for_y.values
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```
